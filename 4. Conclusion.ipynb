```python
# To-Do create a summary about each cluster : 
# percentages of each target (actual Category) within each cluster 
# + Percentages of each predicted class within the cluster 
# results should be something like : 
# For clusterk : {'true_class0_percetage': 0.0,
#                 'true_class1_percetage': 0.0,
#                  'true_class2_percetage': 0.0,
#                   'true_class3_percetage': 0.0,
#                   'true_class4_percetage': 0.0,
#                   'predicted_class0_percetage': 0.0,
#                   'predicted_class1_percetage': 0.0,
#                   'predicted_class2_percetage': 0.0,
#                   'predicted_class3_percetage': 0.0,
#                    'predicted_class4_percetage': 0.0}
```

```python
df1=df.drop(['ID',"Age","Sex","ALB","ALP","ALT","AST","BIL","CHE","CHOL","CREA","GGT","PROT"], axis=1)
df1.shape
```

```python
grouped_df=df1.groupby(['cluster','Category']).size().reset_index(name='count')
grouped_df2=df1.groupby(['cluster','Predicted_class']).size().reset_index(name='count')
print(grouped_df)
print(grouped_df2)
df2=df1.groupby(['Category']).size()
df3=df1.groupby(['Predicted_class']).size()
print(df2)
print(df3)
def percentage_calculate(cls):
    for i in grouped_df.groupby(['cluster']).get_group(cls).index:
        for j in range (0,5): 
            if(grouped_df.groupby(['cluster']).get_group(cls)["Category"][i]==j):
                print("true_class",j,"_percentage:",(grouped_df.groupby(['cluster']).get_group(cls)["count"][i]*100)/df2[j],"% ,")
    
    for i in grouped_df2.groupby(['cluster']).get_group(cls).index:
        for j in range (0,5):
            if(grouped_df2.groupby(['cluster']).get_group(cls)["Predicted_class"][i]==j):
                print("predicted_class",j,"_percentage:",(grouped_df2.groupby(['cluster']).get_group(cls)["count"][i]*100)/df3[j],"% ,")
```

```python
print("For cluster",0,": {")
percentage_calculate(0)
print("}")
print()

print("For cluster",1,": {")
percentage_calculate(1)
print("}")
print()     


print("For cluster",2,": {")
percentage_calculate(2)  
print("}")
print()          


print("For cluster",3,": {")
percentage_calculate(3)
print("}")
print()

print("For cluster",4,": {")
percentage_calculate(4)
print("}")
print()
```
